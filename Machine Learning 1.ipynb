1) Please plot the noisy data and the polynomial you found (in the same figure). Please use
polynomial order of m = 1, 2, 3, 4, 5, 6, 7, 8, respectively

```python
import matplotlib.pyplot as plt             # matplotlib.pyplot used to create plots and visualization
import numpy as np

noise_scale = 100
number_of_samples = 50
x = 30 * (np.random.rand(number_of_samples, 1) - 0.5)
y = 2 * x + 11 * x**2 + 3 * x**3 + noise_scale * np.random.randn(number_of_samples, 1)


# Scatter plot of the original data
plt.scatter(x, y, c='Blue')


x = x.ravel()                    # flatten x array

# Plot the polynomial fits for various degrees
for degree in range(1, 9):
    # Polynomial fit
    coeff = np.polyfit(x, y, deg=degree)
    
    # Generate yy using the polynomial coefficients and xx
    xx = np.linspace(min(x), max(x), 100)
    yy = np.polyval(coeff, xx)  # Evaluate the polynomial
    
    # Plot the curve
    plt.plot(xx, yy, label=f'Degree {degree}')

#print title for graph
plt.title('Noisy Data and Polynomials of Different Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

2. Plot MSE versus order m, for m = 1, 2, 3, 4, 5, 6, 7, 8 respectively. Identify the best choice of m.

```python
import matplotlib.pyplot as plt
import numpy as np

noise_scale = 100
number_of_samples = 50
x = 30*(np.random.rand(number_of_samples, 1) - 0.5)
y = 2 * x + 11 * x**2 + 3 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
# plt.plot(x,y,'ro')


# Scatter plot of the original data
# plt.scatter(x, y, c='Blue', label='Noisy Data')

x = x.ravel()
# Initialize an array to store MSE values for each degree
mse_values = []

# Plot the polynomial fits for various degrees and calculate MSE
for degree in range(1, 9):
    # Polynomial fit
    coeff = np.polyfit(x, y, deg=degree)
    
    # Generate 'yy' using the polynomial coefficients and 'x'
    yy = np.polyval(coeff, x)  # Evaluate the polynomial
    
    # Calculate Mean Squared Error (MSE)
    mse = np.mean((y - yy)**2)
    mse_values.append(mse)
    
    # Plot the curve
    
#     plt.plot(x, yy, label=f'Degree {degree}')

# Plot Mean Squared Error (MSE) versus degree
plt.figure()
plt.plot(range(1, 9), mse_values, marker='o')
plt.xlabel('Degree (m)')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Mean Squared Error (MSE) vs. Polynomial Degree')

# Find the best choice of m (degree) with the minimum MSE
best_degree = np.argmin(mse_values) + 1  # Add 1 because degrees start from 1
plt.axvline(x=best_degree, color='r', linestyle='--', label=f'Best Degree: {best_degree}')
plt.legend()

# Display the plots

plt.show()

print(f'The best choice of m (degree) is: {best_degree}')
```
